```python
dataP=pd.read_csv("http://hiperc.buffalostate.edu/courses/M61xC512/uploads/Site/tree-classification.csv")
dataP
```

```python
def classify(row):
    ret = 0
    if row[1] > 160:
        if row[3] > 3:
            if row[2] < 10:
                ret = 510
            else:
                ret = 382
        else:
            if row[2] < 40:
                ret = 335
            else:
                ret = 312
    else:
        if row[2] < 25:
            if row[3] > 2:
                ret = 341
            else:
                ret = 258
        else:
            if row[1] > 100:
                ret = 285
            else:
                ret = 217
    return ret
        
```

```python
for row in data5:
    row.append(classify(row))
data5
```
